# Tirando conclusões usando query

```python
# Carregue o arquivo `winequality_edited.csv`
import pandas as pd
import numpy as np

df_wine = pd.read_csv('winequality_edited.csv')
df_wine.head()
```

```python
# Printando a coluna alcohol
df_wine['alcohol'].describe()
```

### Será que vinhos com maior teor alcóolico recebem avaliações melhores?

```python
# Obtendo o valor mediano do teor alcóolico
alcohol_median = df_wine['alcohol'].median()
alcohol_median
```

```python
# Obtendo a mediana de df_wine
df_wine.median()
```

```python
# Selecionando amostras com teor alcóolico abaixo da mediana
low_alcohol = df_wine.query('alcohol < @alcohol_median')
low_alcohol.head(10)
```

```python
# Selecionando amostras com teor alcóolico maior ou igual à mediana
high_alcohol = df_wine.query('alcohol >= @alcohol_median')
high_alcohol.head(10)
```

```python
# Certificando que estas consultas incluíram cada amostra uma única vez
num_samples = df_wine.shape[0]
num_samples == low_alcohol['quality'].count() + high_alcohol['quality'].count() # resultado deve ser True
```

```python
# Obter a avaliação média de qualidade para grupos com alto e baixo teor alcóolico
print(" Avaliação média para alto teor:{0} \n Avaliação média para baixo teor:{1}".format(
        high_alcohol["quality"].mean(), low_alcohol["quality"].mean() ))
```

### Vinhos mais suaves recebem avaliações melhores?

```python
# Obter o valor mediano do nível de açúcar residual
residual_median = df_wine['residual_sugar'].median()
residual_median
```

```python
# Selecionar amostras com nível de açúcar residual abaixo da mediana
low_sugar = df_wine.query('residual_sugar < @residual_median')
low_sugar.head(10)
```

```python
# selecione amostras com nível de açúcar residual maior ou igual à mediana
high_sugar = df_wine.query('residual_sugar >= @residual_median')
high_sugar.head(10)
```

```python
# Certificar que estas consultas incluíram cada amostra uma única vez
num_samples == low_sugar['quality'].count() + high_sugar['quality'].count() # resultado deve ser True
```

```python
# Obter a avaliação média de qualidade para grupos com alto e baixo nível de açúcar residual
print(" Avaliação média para alto nível:{0} \n Avaliação média para baixo nivel:{1}".format(high_sugar['quality'].mean(), low_sugar['quality'].mean()))
```
