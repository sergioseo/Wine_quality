# Anexando dados
Primeiro vamos importar os pacotes necessários e carregar os arquivos `winequality-red.csv` e `winequality-white.csv`.

```python
# importe o numpy e o pandas
import numpy as np
import pandas as pd
```

```python
# Vamos carregar os conjuntos de dados de vinhos tintos e brancos
red_df = pd.read_csv('winequality-red.csv', sep=';')
red_df.head()
```

```python
# Vamos renomear a Column para ficar no mesmo padrão
red_df.rename(columns = {"total_sulfur-dioxide": "total_sulfur_dioxide"}, inplace="True")
red_df.head()
```

```python
# Carregando o dataset 'winequality-white.csv'
white_df = pd.read_csv('winequality-white.csv', sep=';')
white_df.head()
```

## Crie colunas de cor
Vamos criar dois vetores de tamanho igual ao número de linhas nos dataframes tinto e branco que repetir o valor “red” ou “white”. O NumPy oferece uma forma bem fácil de fazer isso.

```python
# Criando vetor de cor para o dataframe tinto
color_red = np.repeat("red", red_df.shape[0])
print(color_red)
```

```python
# Criando vetor de cor para o dataframe branco
color_white = np.repeat("white", white_df.shape[0])
print(color_white)
```

* Adicionar os vetores de cor aos dataframes tinto e branco. Vamos fazê-lo associando uma nova coluna chamada 'color' ao vetor apropriado. A célula abaixo faz isso para o dataframe tinto.

```python
red_df['color'] = color_red
red_df.head()
```

* Vamos fazer o mesmo para o dataframe branco e usar `head()` para confirmar a mudança.

```python
white_df['color'] = color_white
white_df.head()
```

## Combinar dataframes com append

```python
# Anexando dataframes
wine_df = red_df.append(white_df)

# Exibindo o dataframe para ver se tudo deu certo
wine_df.head()
```

## Salvando o dataset combinado
Vamos salvar o dataframe recém-combinado como `winequality_edited.csv`. Lembrar de configurar `index=False` para evitar salvar com uma coluna não-nomeada.

```python
wine_df.to_csv("winequality_edited.csv", index=False)
```
